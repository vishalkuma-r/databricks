#Business Transformation and Modeling

```python
query = """
select Row_number() over(order by ci.customer_id) as customer_key,
ci.customer_id,
ci.customer_firstname,
ci.customer_lastname

from silver.customer ci
"""


df=spark.sql(query)
```

Write it to Gold Table

```python
(
    df
    .write
    .format("delta")
    .mode("overwrite")
    .saveAsTable("gold.dim_customers")
)
```
